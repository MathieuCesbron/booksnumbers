# Imports

```python
import json
import matplotlib.pyplot as plt
```

# Get Data

```python
data_json_path = "./data.json"
numbers = {}
with open(data_json_path, "r") as file:
    data = json.loads(file.read())

    numbers = {int(number): occurence for number, occurence in data["stats"].items()}
```

# Plot occurence of numbers

```python
# Extracting values for keys 0 to the limit
limit = 10
values = [numbers.get(key, 0) for key in range(1, limit)]

# Keys ranging from 0 to limit
keys = list(range(1, limit))
colors = "gray"

plt.bar(keys, values, color=colors)

plt.xticks(range(limit))
plt.xlim(0.5, limit - 0.5)

plt.title('Counts of each numbers 1-9 in diverse books', fontsize=16, fontweight="bold", y=1.1)
plt.xlabel('Numbers')
plt.ylabel('Counts')

plt.show()
```
